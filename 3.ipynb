```
# Формирование одного столбца из двух
df['cat_name'] = df['major_cat_name'] + ' | ' + df['minor_cat_name']
df['act_name'] = df['primary_act_name'] + ' | ' + df['secondary_act_name']
# Удаление столбцов одинаковых по смыслу с теми, что уже есть в таблице
del_columns = ['secondary_act_id', 'primary_act_name', 'secondary_act_name', 'major_cat_name', 'minor_cat_name', 'sales_ord_tran_dt', 'edu_1st_indv_val', 'edu_2nd_indv_val', 'occpn_1st_val', 'occpn_2nd_val']
for i in del_columns:
  del df[i]

df.shape
```
