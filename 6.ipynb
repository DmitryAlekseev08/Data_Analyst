**В заданиях подобным этому, не забудьте о том как действует цена в этой базе: она действует от price.start_date до price.end_date, причем если price.end_date IS NULL, то цена действует до сих пор!**

```
# Чтобы показать что если price.end_date IS NULL, то цена действует до сих пор, я использовал COALESCE(end_date, Now())
pd.read_sql_query("""
SELECT description,
       ship_date,
       actual_price
FROM   price
       INNER JOIN product
               ON price.product_id = product.product_id
       INNER JOIN item
               ON product.product_id = item.product_id
       INNER JOIN sales_order
               ON item.order_id = sales_order.order_id
WHERE  ( ( ship_date BETWEEN start_date AND COALESCE(end_date, Now()) )
         AND ( actual_price < ( 0.75 * list_price ) ) ) 
                              """, conn)
```

### Задание 35

**Скидка здесь это list_price*quantity-item.total**

```
pd.read_sql_query("""
SELECT Sum(list_price * quantity - item.total) AS
       "Total amount of discounts granted to the STADIUM SPORTS in 1989"
FROM   price
       INNER JOIN product
               ON price.product_id = product.product_id
       INNER JOIN item
               ON product.product_id = item.product_id
       INNER JOIN sales_order
               ON item.order_id = sales_order.order_id
       INNER JOIN customer
               ON sales_order.customer_id = customer.customer_id
WHERE  ( NAME = 'stadium sports' )
       AND ( ship_date BETWEEN '1989-01-01' AND '1989-12-31' ) 
               """, conn)
```

### Задание 40
Выбрать названия товаров, для которых нынешняя цена увеличилась по сравнению с ценой на 15 декабря 1989г. более, чем на 15%.

1. Получите цену на 15 декабря 1989г.
2. Получите цену, действующую сейчас

Сравнивать тут нужно только list_price.

```
pd.read_sql_query("""
SELECT DISTINCT description
FROM   price
       INNER JOIN product
               ON price.product_id = product.product_id
WHERE  ( Now() BETWEEN start_date AND COALESCE(end_date, Now()) )
       AND ( list_price / (SELECT list_price
                           FROM   price
                           WHERE  ( '1989-12-15' BETWEEN start_date AND COALESCE(end_date, Now()) )
                                  AND ( price.product_id = product.product_id ))) > 1.15 
               """, conn)
```
